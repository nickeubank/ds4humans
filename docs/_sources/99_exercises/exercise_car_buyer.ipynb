# Kelly Blue Book Buyer





In front of you are two people interested in selling their Toyota Camry's. Below are the prices your team have estimated as the likely resale price for cars from different years, with different mileages, and in different conditions. These numbers are obviously not the exact models one gets if you go to Kelly Blue Book, but for the purpose of this exercise, you may assume that these are the output of a highly-refined machine learning algorithm.

Ask the two people selling used cars about their cars' condition, model year, and mileage. They should have all this information on hand, including the car's condition (which they determined by filling out the questionnaire on the Kelly Bluebook website). 

**Offer them a price for their car.** Use the tables below to figure out the likely value of the car, then subtract 10% from the price to make your offer (so you can make a profit when you sell the car).

**BECAUSE YOU ARE PRETENDING TO BE THE WEBSITE ALGORITHM, YOU MAY NOT ASK FOR THINGS BEYOND THE CONDITION (based on four categories), MILEAGE, and MODEL YEAR.**
